```python
import tkinter as tk
from tkinter import *
from tkinter.ttk import *
from tkinter import filedialog
from PIL import ImageTk, Image
from tkinter.scrolledtext import ScrolledText
import subprocess
from tkinter import messagebox
import shutil
import nbformat
from nbconvert.preprocessors import ExecutePreprocessor
from nbconvert import HTMLExporter
from IPython.display import display, HTML
import os
import webbrowser

root = tk.Tk()

root.geometry("800x500")
root.configure(bg='#151515')
root.title("Brain Tumor Detection System")


label1 = tk.Label(root, text="Brain Tumor Detection System",fg='#ffffff', font=('Arial', 18,'bold'), bg='#fa8842')
label1.pack(padx=20, pady=20, ipadx=350)

label2 = tk.Label(root, text="Hello and welcome to Our GUI. Please select the images you want to diagnose.\n\n", fg='#ffffff',font=('Arial', 10), bg='#151515')
label2.pack(anchor="w",padx=20, pady=20, ipadx=350)

frame = tk.Frame(root, bg='#151515', height=800, width=500)
frame.pack(anchor='w')

btn_browse = tk.Button(frame, text='Select Image', bg='#fa8842', fg='#ffffff', font=('verdana', 12))

btn_browsetwo = tk.Button(frame, text='Diagnose', bg='#fa8842', fg='#ffffff', font=('verdana', 12))

image_frame = tk.Frame(root, bg='#151515')
image_frame.pack()



def select_images():
    file_paths = filedialog.askopenfilenames(initialdir="/", title="Select Images",
                                            filetypes=(("Image files", "*.jpg *.png"), ("All files", "*.*")))
    if file_paths:
        display_images(file_paths)
        save_image(file_paths)

def execute_ipynb_file():
    file_path = r"C:\Users\USER\GUI Prediction Model.ipynb"
    execute_notebook(file_path)
        
def execute_notebook(file_path):
    # Read the IPython notebook file
    with open(file_path, 'r') as f:
        nb = nbformat.read(f, as_version=4)

    # Create an ExecutePreprocessor instance
    executor = ExecutePreprocessor(timeout=None)

    # Execute the notebook
    try:
        executor.preprocess(nb)
    except Exception as e:
        messagebox.showerror("Error", str(e))
        return

    # Convert the executed notebook to HTML
    exporter = HTMLExporter()
    html, _ = exporter.from_notebook_node(nb)

    # Save the HTML output to a file
    output_file_path = r"C:\Users\USER\Desktop\Test Run\diagnosis_output.html"
    with open(output_file_path, 'w', encoding='utf-8') as f:
        f.write(html)

    # Open the HTML file in the default web browser
    webbrowser.open_new_tab(f'file://{output_file_path}')
        
    # Display the HTML output on a new window
    output_window = tk.Toplevel(root)
    output_window.title("Diagnosis Output")

    output_frame = Frame(output_window)
    output_frame.pack(fill='both', expand=True)

    output_text = ScrolledText(output_frame, wrap='word')
    output_text.pack(fill='both', expand=True)
    output_text.insert('1.0', html)

    # Disable text editing
    output_text.configure(state='disabled')

    # Add a "Close" button to close the output window
    close_button = Button(output_window, text='Close', command=output_window.destroy)
    close_button.pack(pady=10)

def save_image(file_paths):
    # Set the destination folder path
    folder_path = "C:/Users/USER/Desktop/images taken"  # Replace with the desired folder path

    # Copy each image to the destination folder
    for file_path in file_paths:
        shutil.copy(file_path, folder_path)


def display_images(image_paths):
    num_images = len(image_paths)
    rows = 2
    columns = (num_images + 1) // rows

    # Clear the previous images (if any)
    for widget in image_frame.winfo_children():
        widget.destroy()

    # Display the images in a grid layout
    for i, image_path in enumerate(image_paths):
        img = Image.open(image_path)
        img = img.resize((200, 200), Image.ANTIALIAS)
        photo = ImageTk.PhotoImage(img)

        row = i // columns
        column = i % columns

        label = tk.Label(image_frame, image=photo)
        label.image = photo
        label.grid(row=row, column=column, padx=10, pady=10)


btn_browse['command'] = select_images
btn_browsetwo['command'] = execute_ipynb_file
frame.pack(anchor="center")
btn_browse.pack(side='left', padx=10, pady=10)
btn_browsetwo.pack(side='left', padx=10, pady=10)

root.mainloop()
```
